```python
# create base valves:
# 1. closed/ unusued
bv_off = deepcopy(bt[1][1][1])
print('-- Closed valve:', bv_off)

# 2. open - shattered, duty = 11%
bv_sh = deepcopy(bt[1][1][2])
print('-- Shattering valve:', bv_sh)

# 3. open - clean
bv_cl = deepcopy(bt[4][1][2])
bv_cl['shatter_duty'] = 0.99
print('-- Clean valve:', bv_cl)

# 4. virtual trigger valve
bv_trigger = deepcopy(bt[1][1][-1])
print('-- Trigger valve:', bv_trigger)
```

```python
# make different concentrations (pulse duty) from base valve
sd0 = deepcopy(bv_off)
sd11 = deepcopy(bv_sh)
print('11%:', sd11['shatter_duty'])
sd22 = deepcopy(sd11)
sd22['shatter_duty'] = 0.22
print('22%:', sd22['shatter_duty'])
sd46 = deepcopy(sd11)
sd46['shatter_duty'] = 0.46
print('46%:', sd46['shatter_duty'])
sd100 = deepcopy(bv_cl)
print('100%:clean', sd100['isClean'], ', shatter', sd100['isShatter'])
```

```python
# make base trial (all odor valves closed by trigger valve on)
bt_off = []
for i in range(8):
    bt_off.append(bv_off)
bt_off.append(bv_trigger)

print(len(bt_off))
print('example valve:', bt_off[1])
```

```python
# write a function to check valve details
def printValvePara_singleTrial(aTrial):
    """print [length, isShatter, shatter_duty, isClean, offset] for each valve in trial"""
    for i in range(aTrial[0]):
        print(aTrial[2], 'valve', i, 'length:', aTrial[1][i]['length'],
            ', shatter:', aTrial[1][i]['isShatter'],', shatter_duty', aTrial[1][i]['shatter_duty'],
            ', clean:', aTrial[1][i]['isClean'], ', offset:', aTrial[1][i]['offset'])
    print('\n')
```

```python
# write a function to check valve details for each trial in a trial bank
def printValvePara(aTrialBank):
    """print [length, isShatter, shatter_duty, isClean, offset] for each valve in trial"""
    print(f'This trial bank has {len(aTrialBank)} trials.')
    for n in range(len(aTrialBank)):
        print(f'For trial idx {n}:')
        for i in range(aTrialBank[n][0]):
            print(aTrialBank[n][2], 'valve', i, 'length:', aTrialBank[n][1][i]['length'],
                ', shatter:', aTrialBank[n][1][i]['isShatter'],', shatter_duty', aTrialBank[n][1][i]['shatter_duty'],
                ', clean:', aTrialBank[n][1][i]['isClean'], ', offset:', aTrialBank[n][1][i]['offset'])
        print('\n')
```

```python
# randomise trials
tb_ABCDr = deepcopy(tb_ABCD)

random.seed(50)
idx_original = list(range(len(tb_ABCDr)))
idx_shuffle = random.sample(idx_original, len(idx_original))
print(idx_original), print(idx_shuffle)

# !!! need to call random.seed before every random operation!
random.seed(50)
tb_ABCDr = random.sample(tb_ABCDr, len(tb_ABCDr))
printValvePara(tb_ABCDr)
```

```python
# verify the randomising order
expectedTrial = []
resultTrial = []
for i in range(len(idx_shuffle)):
    expectedTrial.append(tb_ABCD[idx_shuffle[i]][2])
    resultTrial.append(tb_ABCDr[i][2])
print('expected trial order:', expectedTrial)
print('resulting trial order', resultTrial)
```

```python
#  chunk function from internet
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
# try this chunk function from internet
pprint.pprint(list(chunks(idx_original, 15)))
pprint.pprint(list(chunks(idx_shuffle, 15)))
```
